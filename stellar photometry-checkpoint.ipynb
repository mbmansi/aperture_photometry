```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.io import fits
from astropy.visualization import simple_norm
```

```python
hdulist = fits.open("./procyon/procyon20ms-0022G.fit")
# hdulist.info()
img_original = fits.getdata("./procyon/procyon20ms-0022G.fit")
```

```python
norm = simple_norm(img_original, 'log', log_a=1000)
```

```python
y_max, x_max=(np.where(img_original==np.max(img_original)))
y_max=int(y_max)
x_max=int(x_max)
print(y_max, x_max)
#
```

```python
plt.imshow(img_original, cmap='Greys_r', origin='lower', norm=norm)
#plt.scatter(y_max, x_max, fc='none', ec='k', s=400)
lims = 400
plt.ylim(y_max-lims, y_max+lims)
plt.xlim(x_max-lims, x_max+lims)
```

```python
# background_radius = 50
img = img_original[y_max-lims:y_max+lims, x_max-lims:x_max+lims]
plt.imshow(img, origin='lower')
center_y, center_x = np.where(img==np.max(img))
center_x = int(np.mean(center_x))
center_y = int(np.mean(center_y))
print(center_x, center_y)
```

```python
distances_array = np.zeros_like(img)
for i in range(0, len(img)):
    for j in range(0, len(img[0])):
        distances_array[i, j] = np.sqrt((i-center_x)**2 + (j-center_y)**2)
```

```python
def create_mask(a):
    return (distances_array<a)*img
```

```python
radii = np.arange(40)
count = np.zeros_like(radii)
area = np.zeros_like(radii)
for r in radii:
    count[r] = np.sum(create_mask(r))
    area[r] = np.sum(distances_array<r)
```

```python
def new_mask(x1, y1, radius):
    temp_dist = np.zeros_like(img)
    x = x1 + center_x
    y = y1 + center_y
    for i in range(0, len(img)):
        for j in range(0, len(img[0])):
            temp_dist[i, j] = np.sqrt((i-x)**2 + (j-y)**2)
    return(temp_dist<radius)

def get_count(x1, y1, radius):
    return(np.sum(new_mask(x1, y1, radius)*img))
```

```python
import random
def get_bg(rad=40):
    n_points=8
    bg=np.zeros(n_points)
    bg_area=np.zeros(n_points)
    for i in range(0,n_points):
        bg[i]=get_count(random.randrange(100,200),random.randrange(100,200),rad)
        bg_area[i]=bg[i]/(40*40*np.pi)
    print(bg_area)
    avg_bg=np.mean(bg_area)    
#             np.append(bg,get_count(dist_star,dist_star,rad))
    return(avg_bg)
back=get_bg() 
print(back)
```
